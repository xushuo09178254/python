```python
import pandas as pd
import numpy as np


df = pd.read_csv(r'C:\Users\26508\Downloads\titanic.csv')

df.head()
```

```python
df_cleaned = df.dropna()

df_cleaned.info()
```

```python
import matplotlib.pyplot as plt  # 仅用于画图
survived_counts = df_cleaned['Survived'].value_counts()
labels = ['未存活', '存活'] if 0 in survived_counts.index else ['存活']
plt.figure(figsize=(6, 6))
plt.pie(survived_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightskyblue'])
plt.title('泰坦尼克号存活比例')
plt.show()
```

```python
df_sorted = df_cleaned.sort_values(by='Age')
plt.figure(figsize=(10, 6))
plt.plot(df_sorted['Age'].values, df_sorted['Fare'].values, color='green')
plt.xlabel('年龄')
plt.ylabel('票价')
plt.title('不同年龄乘客的票价趋势')
plt.grid(True)
plt.show()
```
